```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
path = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
```

```python
data_raw = pd.read_csv(path)
```

```python
data_raw.head()
```

```python
data_raw.tail()
```

```python
is_state = data_raw['Province_State']=='Massachusetts'
data = data_raw[is_state]
```

```python
data.drop(columns=['UID'], inplace=True)
```

```python
data.drop(columns=['iso2','iso3', 'code3', 'FIPS', 'Province_State', 'Country_Region', 'Lat', 'Long_', 'Combined_Key'], inplace=True)
```

```python
data.set_index('Admin2', inplace=True)
```

```python
data.head(15)
```

```python
# Line plot for Counties
data.T.plot(figsize=(16,8))
plt.show()
```

```python
data.T.loc['3/1/20':,:].plot(figsize=(16,6))
plt.show()
```

```python
data.T.loc['3/1/20':,['Barnstable', 'Berkshier', 'Bristol', 'Dukes', 'Essex', 'Franklin']].plot(figsize=(16,6))
plt.show()
```

```python
data_total= data.sum(axis=1)
```

```python
data['Total'] = data_total
```

```python
data.head()
```

```python
top = data.nlargest(8,'Total')
```

```python
top
```

```python
barplot = top.loc[:,'Total'].plot(kind='barh',figsize=(16,8))
for i, v in enumerate(top['Total']):
    barplot.text(v + 3, i, " "+str(v),va='center', color='black',fontweight='bold' )
plt.title('Total Cases for top eight Massachusetts Counties')
plt.xlabel('Total Cases')
plt.ylabel('Counties')

```
